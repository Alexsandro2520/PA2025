```python
#Entrada
renda_hora=float(input("Salário por hora: "))
horas_trabalhadas=float(input("Horas trabalhadas: "))
minutos_trabalhados=float(input("Minutos trabalhadas: "))

#Processamento
horas_trabalhadas=horas_trabalhadas+((minutos_trabalhados*(100/60))/100)
lucro=horas_trabalhadas*renda_hora
lucro_total=round(lucro+(lucro*0.7), 2)

#Saida
print(f"Então você ganha {lucro_total}!")
```
