##1 - Número Positivo ou Negativo

```python
print("-"*50, "Número positivo ou negativo", "-"*50, "\n\n")

num=int(input("Digite um número, que eu le direi se é negativo ou positivo: "))

if(num>0):
  print("É positivo rsrs")
elif(num<0):
  print("É negativo")
else:
  print("É zero meu mano.")
```

##2 - Par ou Ímpar

```python
print("-"*50, "Par ou Ímpar", "-"*50, "\n\n")

num=int(input("Me diga um numero e eu le direi se é par ou ímpar: "))

if(num/2==int(num/2)):
  print(f"{num} é par")
else:
  print(f"{num} é ímpar")
```

##3 - Verificação de maior de idade

```python
print("-"*50, "Vericação de maior de idade", "-"*50, "\n\n")

num=int(input("Me diga tua idade e eu le direi se tu é maior de idade ou não: "))

if(num>18):
  print("Tu é maior de idade!")
elif(num<18 and num>0):
  print("Tu é menor de idade!")
else:
  print("Tu não existe!!!!")
```

##8 - Classificações de um Triangulo

```python
print("-"*50, "Classificaçõs de um triangulo", "-"*50, "\n\n")

l1=int(input("Lado 1: "))
l2=int(input("Lado 2: "))
l3=int(input("Lado 3: "))

isequilatero=(l1==l2==l3)
isisoceles=(l1==l2) or (l2==l3) or (l1==l3)
isescaleno=(l1!=l2!=l3)

if(isequilatero):
  print("Seu triangulo é um equilatero")
elif(isisoceles):
  print("Seu triangulo é um isóceles")
elif(isescaleno):
  print("Seu triangulo é um escaleno")
```
